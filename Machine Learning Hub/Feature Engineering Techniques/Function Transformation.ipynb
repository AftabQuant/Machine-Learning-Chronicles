```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"D:\Ultimate Programming\Data Bases\Cleaned Dataset\loan_data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
q1 = df['installment'].quantile(0.25)
q3 = df['installment'].quantile(0.75)

iqr = q3-q1
min_range = q1 - 1.5*iqr
max_range = q3 + 1.5*iqr
df = df[df['installment']<max_range]
```

```python
sns.distplot(df['installment'])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer

ft = FunctionTransformer(func=np.log1p)
ft.fit(df[['installment']])
```

```python
df['installment_ft'] = ft.transform(df[['installment']])
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
sns.distplot(df['installment'])

plt.subplot(1,2,2)
sns.distplot(df['installment_ft'])
plt.show()
```
