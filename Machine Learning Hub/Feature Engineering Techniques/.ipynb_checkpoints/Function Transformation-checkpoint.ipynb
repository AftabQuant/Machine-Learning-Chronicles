```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel(r"D:\Ultimate Programming\Data Bases\ESD.xlsx")
```

```python
df.head()
```

```python
df.drop(columns=['Exit Date', 'Hire Date', 'Full Name', 'EEID'], axis=1, inplace=True)
```

```python
df.head(2)
```

### Annual Salary Distribution

```python
col = df['Annual Salary']
sns.distplot(col, kde = True)
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
ft = FunctionTransformer(func=np.log1p)
```

```python
ft_col = ft.fit_transform(col)
```

```python
plt.figure(figsize=(13, 5))
plt.subplot(1,2,1)
sns.distplot(col, kde = True, color='Red')
plt.title('Befor Func Transformation')
plt.grid()

plt.subplot(1,2,2)
sns.distplot(ft_col, kde = True, color='Blue')
plt.title('After Func Transformation')

plt.grid()
plt.show()
```
