```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel(r"D:\Ultimate Programming\Data Bases\Banking Data.xlsx")
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df.info()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(x=df['Credit Card Balance'])

plt.subplot(1,2,2)
sns.boxplot(x=df['Bank Loans'])
plt.show()
```

```python
df.describe()
```

```python
q1 = df['Credit Card Balance'].quantile(.25)
q3 = df['Credit Card Balance'].quantile(.75)

iqr = q3- q1
min_range = q1 - 1.5*iqr
max_range = q3 + 1.5*iqr
df = df[df['Credit Card Balance']<max_range]
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
for col in df.select_dtypes(include='float64').columns:
    df[col] = ss.fit_transform(df[[col]])
```

```python
x = df[['Credit Card Balance']]
y = df[['Bank Loans']]
```

```python
sns.scatterplot(data=df, x='Credit Card Balance', y='Bank Loans')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
df.head(2)
```

```python
lr.predict([[-1.143686]])
```

```python
lr.score(x_test, y_test)*100
```
