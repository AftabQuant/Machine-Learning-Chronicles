```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder

tr = TransactionEncoder()
tr.fit(df)
```

```python
tr.transform(df)
tr.columns_
```

```python
df_tr = pd.DataFrame(tr.transform(df), columns=tr.columns_)
df_tr.head(3)
```

```python
df_tr.shape
```

```python
from mlxtend.frequent_patterns import apriori

apriori(df_tr, min_support=0.00001, use_colnames=True, max_len=3).sort_values(by='support')
```
