```python
df['Month Name'] = df['Hire Date'].dt.strftime('%b')
```

```python
month_table = df['Month Name'].value_counts().sort_index().rename_axis('Month').reset_index(name='count')
month_table['Month'] = month_table['Month'].sort_values()
```

```python
month_table
```

```python
plt.plot(month_table['Month'], month_table['count'])
plt.show()
```
