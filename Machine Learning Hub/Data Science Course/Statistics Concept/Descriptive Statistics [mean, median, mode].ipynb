```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"D:\Ultimate Programming\Data Bases\ESD.xlsx")
df.head(4)
```

```python
df.drop(columns=["EEID", "Exit Date", "Bonus %"], inplace=True, axis=1)
```

```python
df.head(4)
```

```python
mn = df["Annual Salary"].mean()
```

```python
md = df["Annual Salary"].median()
```

```python
me = df["Annual Salary"].mode()[0]
```

```python
print(mn, md, me)
```

```python
plt.figure(figsize=(10,7))
sns.histplot(x="Annual Salary", data=df)
plt.plot([mn for i in range(0, 200)], [i for i in range(0, 200)], c="green", label="mean")
plt.plot([md for i in range(0, 200)], [i for i in range(0, 200)], c="red", label="median")
plt.plot([me for i in range(0, 200)], [i for i in range(0, 200)], c="yellow", label="mode")

plt.legend()
plt.show()
```
