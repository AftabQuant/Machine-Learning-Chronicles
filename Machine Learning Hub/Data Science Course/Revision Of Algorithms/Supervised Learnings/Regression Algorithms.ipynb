```python
x = df.drop('AI_Readiness_Score', axis=1)
y = df['AI_Readiness_Score']
```

# Data Preprocessing

```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer(
    transformers=[
        ('si', SimpleImputer(), [0, 2, 3, 4, 5, 6, 7, 8, 9, 10]),
        ('ohe', OneHotEncoder(drop='first', sparse_output=False), [1]),
        ('ms', MinMaxScaler(), [0, 2, 3, 4, 5, 6, 7, 8, 9, 10])
    ], remainder='passthrough')
```

```python
ct.fit(x)
```

```python
col = ct.get_feature_names_out()
x_en = ct.transform(x)
x_en = pd.DataFrame(x_en, columns=col)
```

```python
x_en.head(4)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_en, y, test_size=0.2, random_state=42)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train)*100
```
