```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"D:\Ultimate Programming\Data Bases\Home Loan DS.csv")
df.head(3)
```

```python
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)
df['Dependents'] = df['Dependents'].replace('3+', '3')
df['Dependents'] = df['Dependents'].astype(np.int32)
```

```python
df.info()
```

```python
df = df.drop('Loan_ID', axis = 1)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
x = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']
```

```python
num = ['Dependents', 'LoanAmount', 'Credit_History']
cat = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area']

ct = ColumnTransformer(
    transformers=[
        ('s', SimpleImputer(), num),
        ('o', OneHotEncoder(drop='first'), cat)
    ], remainder='passthrough'
)
```

```python
ct.fit(x)
```

```python
x_en = ct.transform(x)
```

```python
x_en = pd.DataFrame(x_en, columns=ct.get_feature_names_out())
```

```python
x_en.shape
```

```python
x_en.head(2)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_en, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(class_weight='balanced', random_state=42)
```

```python
dt.fit(x_train, y_train)
```

```python
dt.score(x_train, y_train)*100, dt.score(x_test, y_test)*100
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
params = {
    'criterion' : ["gini", "entropy", "log_loss"],
    'splitter' : ["best", "random"],
    'max_depth' : [i for i in range(2, 11)],
    'class_weight' : ['balanced'],
     'random_state' : [i for i in range(2, 42)]
}
```

```python
gd = GridSearchCV(DecisionTreeClassifier(), params)
gd.fit(x_train, y_train)
```

```python
rd = RandomizedSearchCV(DecisionTreeClassifier(), params, n_iter=50)
rd.fit(x_train, y_train)
```

```python
print(gd.best_params_)
print(gd.best_score_ * 100)
```

```python
print(rd.best_params_)
print(rd.best_score_ * 100)
```
