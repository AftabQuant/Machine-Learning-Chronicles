```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"D:\Ultimate Programming\Data Bases\Home Loan DS.csv")
df.head(3)
```

```python
df.info()
```

```python
data = df.drop(['Loan_Status', 'Loan_ID'], axis=1)
```

```python
data.head(2)
```

```python
sns.pairplot(data=df, hue='Loan_Status')
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
data.info()
```

```python
x = data
y = df['Loan_Status']
```

```python
ct = ColumnTransformer(
    transformers=[
        ('', SimpleImputer(), [6, 7, 8, 9]),
        ('o', OneHotEncoder(drop='first', sparse_output=False), [0, 1, 2, 3, 4, 10])
],remainder='passthrough')
```

```python
ct.fit(x)
```

```python
x = ct.transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
print(f"Training Accuracy: {round(lr.score(x_train, y_train)*100, 2)}")
print(f"Testing Accuracy: {round(lr.score(x_test, y_test)*100, 2)}")
```

```python
lrcv = LogisticRegressionCV()
```

```python
lrcv.fit(x_train, y_train)
```
