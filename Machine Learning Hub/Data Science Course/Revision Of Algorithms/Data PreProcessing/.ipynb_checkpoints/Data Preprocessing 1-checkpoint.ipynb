## Data Perprocessing

```python
data = df.drop('Sales_Classification', axis=1)
data.head(1)
```

```python
data.info()
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
ct = ColumnTransformer(
    transformers=[
       ('si', SimpleImputer(), [1, 6, 7, 8, 9]),
       ('ohe', OneHotEncoder(drop='first', sparse_output=False), [0, 2, 3, 4, 5]),
       ('ms', MinMaxScaler(), [1, 6, 7 ,8 ,9])
    ], remainder='passthrough')
```

```python
ct.fit(data)
```

```python
x_en = ct.transform(data)
```

```python
col = ct.get_feature_names_out()
```

```python
x_en.shape
```

```python
x_en = pd.DataFrame(x_en, columns=col)
x_en.head(2)
```
