# Handle Outlier 

```python
df.describe()
```

```python
sns.boxenplot(data=df, x = 'stays_in_week_nights')
plt.show()
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
df.shape
```

```python
q1 = df['stays_in_week_nights'].quantile(0.25)
q3 = df['stays_in_week_nights'].quantile(0.75)
print(q1, q3)
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (IQR*1.5)
max_range = q3 + (IQR*1.5)
print(min_range, max_range)
```

```python
new_df = df[df['stays_in_week_nights']<=max_range]
```

```python
new_df.shape
```

```python
sns.boxenplot(data=df, x = 'stays_in_week_nights')
plt.show()
```

# Using Z Score

```python
min_range = df['stays_in_week_nights'].mean() - (3*df['stays_in_week_nights'].std())
max_range = df['stays_in_week_nights'].mean() + (3*df['stays_in_week_nights'].std())
```
