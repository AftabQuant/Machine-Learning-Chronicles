```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Grocery Products Purchase.csv')
df.head()
```

```python
df.shape
```

```python
market = []

for i in range(df.shape[0]):
    cus = []
    for j in df.columns:
        if type(df[j][i]) is str:
            cus.append(df[j][i])
    market.append(cus)
```

```python
items = []
for i in market:
    for j in i:
        items.append(j)
```

```python
import collections

data = collections.Counter(items)
```

```python
df = pd.DataFrame({'items': data.keys(), 'values':data.values()})
```

```python
df = df.sort_values(by='values', ascending=False)
```

```python
df.head(6)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
new_df = pd.DataFrame(tr.transform(market), columns=tr.columns_)
```

```python
new_df.head(4)
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(new_df,min_support=0.05,max_len=3, use_colnames=True).sort_values(by='support')
```
