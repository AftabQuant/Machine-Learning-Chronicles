```python
import numpy as np
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Grocery Products Purchase.csv")
df.head(4)
```

```python
market = []

for i in range(df.shape[0]) :
    cus = []
    for j in df.columns :
        if type(df[j][i]) is str:
            cus.append(df[j][i])
    market.append(cus)
```

```python
# Convert a single list

nums = []

for i in market :
    for j in i :
        nums.append(j)
```

```python
import collections

data = collections.Counter(nums)
```

```python
dataset = pd.DataFrame({'items': data.keys(), 'values': data.values()})
dataset.head(4)
```

```python
dataset = dataset.sort_values(by="values", ascending=False)
```

```python
dataset.head(4)
```

```python
from mlxtend.preprocessing import TransactionEncoder

tr = TransactionEncoder()
tr.fit(market)
```

```python
new_df = pd.DataFrame(tr.transform(market), columns=tr.columns_)
new_df.head(2)
```

# Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(new_df,min_support=0.05,max_len=3, use_colnames=True).sort_values(by='support')
```

# F P Growth Algorithm

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fpgrowth(new_df, min_support=0.05, use_colnames=True, max_len=3).sort_values(by='support')
```
