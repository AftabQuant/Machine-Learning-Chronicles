```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
df.head(3)
```

# Hierarchical Clustering

```python
# To print dentogram

import scipy.cluster.hierarchy as sc
```

```python
sc.dendrogram(sc.linkage(df, method='complete', metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

ag = AgglomerativeClustering(n_clusters=3, linkage='single')
ag.fit(df)
```

```python
df['target'] = ag.fit_predict(df)
```

```python
df.head(2)
```

```python
sns.pairplot(data=df, hue='target')
plt.show()
```

```python
from sklearn.cluster import KMeans

knn = KMeans()
knn.fit(df)
```

```python
wscc = []

for i in range(1, 21):
    kn = KMeans(n_clusters=i)
    kn.fit(df)
    wscc.append(kn.inertia_)
```

```python
for i in wscc:
    print(i, end=", ")
```

```python
plt.plot([i for i in range(1,21)], wscc, marker='o')
plt.xticks([i for i in range(1,21)])
plt.xlabel('no of clusters')
plt.ylabel('wscc')
plt.grid(visible=True)
plt.show()
```

```python
knn = KMeans(n_clusters=2)
knn.fit(df)
```

```python
df['target'] = knn.fit_predict(df)
```
