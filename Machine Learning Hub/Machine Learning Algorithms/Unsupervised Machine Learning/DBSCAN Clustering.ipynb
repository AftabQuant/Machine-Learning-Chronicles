```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_moons
```

```python
x, y = make_moons(n_samples=300, noise=0.5)
```

```python
data = {'data1':x[:,0], 'data2':x[:,1], 'output':y}
```

```python
df = pd.DataFrame(data)
df.head(2)
```

```python
sns.scatterplot(data=df, x='data1', y='data2', hue='output')
plt.show()
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
new_df = df.drop(columns='output')
```

```python
new_df.head(3)
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.5)
db.fit(new_df)
```

```python
new_df['target'] = db.fit_predict(new_df)
```

```python
new_df.head(4)
```

```python
df.head(4)
```

```python
plt.figure(figsize=(5,5))
plt.subplot(1,2,1)
plt.title('new dataset')
sns.pairplot(data=new_df, hue='target')

plt.subplot(1,2,2)
sns.pairplot(data=df, hue='output')
plt.title('old dataset')
plt.show()
```
