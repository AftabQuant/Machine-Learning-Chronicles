```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"D:\Ultimate Programming\Data Bases\Machine Learning Datasets\placement-dataset.csv")
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.drop('city', axis=1, inplace=True)
```

```python
df['iq'].fillna(df['iq'].mode()[0], inplace=True)
df['cgpa'].fillna(df['cgpa'].mean(), inplace=True)
```

```python
df.head(2)
```

```python
sns.scatterplot(data=df, x='cgpa', y='iq', hue='placement')
plt.show()
```

```python
x = df.iloc[:, :-1]
y = df['placement']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
sns.distplot(df['cgpa'])
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
#6.8	123.0	1
#5.9	106.0	0

```

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
print(f"GaussianNB: {gnb.score(x_test, y_test)*100:.2f}")
print(f"MultinomialNB: {mnb.score(x_test, y_test)*100:.2f}")
print(f"BernoulliNB: {bnb.score(x_test, y_test)*100:.2f}")
```
