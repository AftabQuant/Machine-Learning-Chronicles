```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"D:\Ultimate Programming\Data Bases\Machine Learning Datasets\Heart Prediction Quantum Dataset.csv")
df.head(4)
```

```python
df = df.drop(columns='HeartDisease')
```

```python
x = df[['HeartRate']]
y = df['QuantumPatternFeature']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVR

sv = SVR()
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
sns.scatterplot(data=df, x='HeartRate', y='QuantumPatternFeature')
plt.plot(df['HeartRate'], sv.predict(x), color='red')
plt.show()
```
