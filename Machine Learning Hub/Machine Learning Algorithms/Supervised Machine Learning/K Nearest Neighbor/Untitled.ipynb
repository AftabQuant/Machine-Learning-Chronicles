```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"D:\Ultimate Programming\Data Bases\Machine Learning Datasets\Heart Prediction Quantum Dataset.csv")
df.head(4)
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:, :-2]
y = df['QuantumPatternFeature']
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x), columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
print(knn.score(x_train, y_train)*100, knn.score(x_test, y_test)*100)
```

```python
for i in range(1, 25):
    kn = KNeighborsRegressor(n_neighbors=i)
    kn.fit(x_train, y_train)
    print(i, knn.score(x_train, y_train)*100, knn.score(x_test, y_test)*100)
```
