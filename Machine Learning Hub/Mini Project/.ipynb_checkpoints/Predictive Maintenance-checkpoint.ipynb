```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"D:\Ultimate Programming\Data Bases\Machine Learning Datasets\Classification\predictive_maintenance.csv")
```

```python
df.head(2)
```

```python
df['Target'].value_counts()
```

```python
df.drop(['Product ID', 'Failure Type', 'UDI', 'Type'], inplace=True, axis=1)
```

```python
df.head(2)
```

```python
x = df.iloc[:, :-1]
y = df['Target']
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
for col in x.select_dtypes(include=['int64', 'float64']).columns:
    x[col] = ss.fit_transform(x[[col]])
```

```python
x.head(2)
```

from imblearn.over_sampling import RandomOverSampler

ro = RandomOverSampler()
ro_x, ro_y = ro.fit_resample(x, y)

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(ro_x, ro_y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

## Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion='entropy')
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)*100
```

```python
plt.figure(figsize=(50, 20))
from sklearn.tree import plot_tree
plot_tree(dt)
plt.show()
```
