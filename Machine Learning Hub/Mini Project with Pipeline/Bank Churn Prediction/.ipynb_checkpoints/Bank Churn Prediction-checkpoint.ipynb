```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"D:\Ultimate Programming\Data Bases\DL DS\Churn_Modelling.csv")
df.head(2)
```

```python
df.drop(columns=['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x = df.drop(columns=['Exited'], axis=1)
y = df['Exited']
```

```python
x.head(2)
```

```python
x.info()
```

```python
trf1 = ColumnTransformer(transformers=[
    ('OneHotEncoding', OneHotEncoder(drop='first', sparse_output=False), [1, 2])
], remainder='passthrough')
```

```python
# x = pd.DataFrame(trf1.fit_transform(x))
```

```python
x.shape
```

```python
trf2 = ColumnTransformer(transformers=[
    ('Standardize', MinMaxScaler(), slice(0, 11))
], remainder='passthrough')
```

```python
# x = pd.DataFrame(trf2.fit_transform(x), columns=x.columns)
```

```python
x.head(2)
```

```python
x.describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr = LogisticRegression()
```

```python
from sklearn.pipeline import make_pipeline
pipeline = make_pipeline(trf1, trf2, lr)
```

```python
pipeline.fit(x_train, y_train)
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
import joblib
joblib.dump(pipeline, 'model.pkl')
```
