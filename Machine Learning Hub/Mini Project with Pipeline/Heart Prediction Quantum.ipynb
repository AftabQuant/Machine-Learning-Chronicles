```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"D:\Ultimate Programming\Data Bases\Machine Learning Datasets\Classification\Heart Prediction Quantum Dataset.csv")
```

```python
df['HeartDisease'].value_counts()
```

```python
x = df.iloc[:, :-1]
y = df['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
print(lr.score(x_test, y_test)*100)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
```

```python
clf = confusion_matrix(y_test, lr.predict(x_test))
clf
```

```python
sns.heatmap(clf, annot=True)
plt.show()
```

```python
print(precision_score(y_test, lr.predict(x_test))*100)
print(recall_score(y_test, lr.predict(x_test))*100)
print(f1_score(y_test, lr.predict(x_test))*100)
```
