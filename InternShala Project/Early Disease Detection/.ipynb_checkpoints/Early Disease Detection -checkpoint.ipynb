```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("D:\\Ultimate Programming\\Data Bases\\Machine Learning Datasets\\Early Disease Detection.xlsx")
df.head()
```

```python
import math
df['ages'] = df['age'].apply(lambda x : math.ceil(x/365))
```

```python
df.head()
df = df.drop(columns='age')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Distribution of target variable

sns.countplot(data=df, x='disease')
plt.title('Distribution Of Disease')
plt.show()
```

```python
# Frequency of disease over age

sns.histplot(data=df,x='ages', hue='disease', multiple='stack', bins=30)
plt.title('Frequency Of Disease Over Age')
plt.show()
```

```python
# Bar plot of age by disease
sns.barplot(data=df, x='ages',y='disease')
plt.title('Bar Plot Of Ages By Disease')
plt.show()
```

```python
# Correlation matrix

num_cols = list()
for col in df.columns:
    if(df[col].dtype != object):
        num_cols.append(col)
corr_matrix = df[num_cols].corr()
plt.figure(figsize=(14,7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()
```
