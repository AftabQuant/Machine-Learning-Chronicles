```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"D:\Ultimate Programming\Data Bases\Machine Learning Datasets\placementdata.csv")
```

```python
df.head(2)
```

```python
df.isnull().sum().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()

for col in df.select_dtypes(include='object').columns:
    df[col] = oe.fit_transform(df[[col]])
```

```python
df.info()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data=df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df['PlacementStatus']
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(x)
```

```python
x = pd.DataFrame(ss.transform(x), columns=x.columns)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
plt.figure(figsize=(8,4))
plt.bar(x.columns, lr.coef_,)
plt.title("Linear Regression")
plt.xlabel("Columns")
plt.ylabel("Coefficient")
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.10)
lasso.fit(x_train, y_train)
```

```python
lasso.score(x_test, y_test)*100 
```

```python
plt.figure(figsize=(8,4))
plt.bar(x.columns, lasso.coef_)
plt.title("Lasso Regularization")
plt.xlabel("Columns")
plt.ylabel("Coefficient")
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=10)
ridge.fit(x_train, y_train)
```

```python
ridge.score(x_test, y_test)*100 
```

```python
plt.figure(figsize=(8,4))
plt.bar(x.columns, ridge.coef_)
plt.title("Ridge Regularization")
plt.xlabel("Columns")
plt.ylabel("Coefficient")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(1,3,1)
plt.figure(figsize=(4,4))
plt.bar(x.columns, lr.coef_,)
plt.title("Linear Regression")
plt.xlabel("Columns")
plt.ylabel("Coefficient")
plt.xticks(rotation=45)

sns.pairplot(1,3,2)
plt.figure(figsize=(4,4))
plt.bar(x.columns, lasso.coef_)
plt.title("Lasso Regularization")
plt.xlabel("Columns")
plt.ylabel("Coefficient")
plt.xticks(rotation=45)

sns.pairplot(1,3,3)
plt.figure(figsize=(4,4))
plt.bar(x.columns, ridge.coef_)
plt.title("Ridge Regularization")
plt.xlabel("Columns")
plt.ylabel("Coefficient")
plt.xticks(rotation=45)
```
